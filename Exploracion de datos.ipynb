## **Exploración de los datos**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pathlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def clean_data_country(root):
    "Nos devuelve el fichero de cada país limpio"
    
    df_country = pd.read_excel(root, header=1)
    df_country = df_country.drop([0], axis=0)
    df_country.rename(columns={'Unnamed: 1':'industry',
                        'Elemento': 'country'}, inplace=True)
    country_name = df_country['country'][1]
    df_country['country'] = df_country['country'].replace(np.nan, country_name)
   
  
    return df_country
```

```python
latam_investment = pd.DataFrame()
```

```python
directory = 'Inversion España en LATAM/'

directory_latam = pathlib.Path(directory)

for file in directory_latam.iterdir():
    country = clean_data_country(file)
    latam_investment = pd.concat([latam_investment, country])
  
```

```python
latam_investment_2 = latam_investment.melt(['industry','country'])
latam_investment_2.head()
```

```python
latam_investment_2 = latam_investment_2[['variable','value','country', 'industry']]
latam_investment_2.head()
```

```python
latam_investment_2.rename(columns={'variable':'year',
                        'value': 'investment'}, inplace=True)
latam_investment_2.head()
```

```python
investment_year_country = latam_investment_2.groupby('year')['country','investment'].sum().reset_index()
investment_year_country.head()
```

```python
plt.figure(figsize=(100,35))
sns.barplot(data=investment_year_country, x='year', y='investment', hue='country')
plt.title('Evolucion de la inversión total por paises')
```
